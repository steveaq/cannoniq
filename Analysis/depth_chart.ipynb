```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch

# Create a pitch object
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='white', stripe=True)
fig, ax = pitch.draw(figsize=(12, 7))

# Define player positions for a 4-3-3 formation (Arsenal's squad)
your_team = {
    'GK': (5, 40),          # Goalkeeper
    'RB': (20, 75),         # Right Back
    'CB1': (20, 50),        # Center Back 1
    'CB2': (20, 30),        # Center Back 2
    'LB': (20, 5),          # Left Back
    'CM1': (35, 60),        # Central Midfielder 1
    'CM2': (30, 40),        # Central Midfielder 2
    'CM3': (35, 20),        # Central Midfielder 3
    'RW': (50, 75),         # Right Winger
    'ST': (50, 40),         # Striker
    'LW': (50, 5)           # Left Winger
}

# Define player positions for a 4-3-3 formation (opposite team - mirrored)
opposite_team = {
    'GK': (115, 40),        # Goalkeeper
    'RB': (100, 75),        # Right Back
    'CB1': (100, 50),       # Center Back 1
    'CB2': (100, 30),       # Center Back 2
    'LB': (100, 5),         # Left Back
    'CM1': (85, 60),        # Central Midfielder 1
    'CM2': (90, 40),        # Central Midfielder 2
    'CM3': (85, 20),        # Central Midfielder 3
    'RW': (70, 75),         # Right Winger
    'ST': (70, 40),         # Striker
    'LW': (70, 5)           # Left Winger
}

# Plot your team (Arsenal) on the pitch
for position, (x, y) in your_team.items():
    pitch.scatter(x, y, ax=ax, color='blue', s=300, edgecolors='black', linewidth=2, label='Arsenal')
    ax.text(x, y, position, color='white', fontsize=12, ha='center', va='center')

# Plot the opposite team on the pitch
for position, (x, y) in opposite_team.items():
    pitch.scatter(x, y, ax=ax, color='red', s=300, edgecolors='black', linewidth=2, label='Opposite Team')
    ax.text(x, y, position, color='white', fontsize=12, ha='center', va='center')

# Add depth chart labels for Arsenal's squad
depth_labels_your_team = {
    'GK': 'Ramsdale',       # Aaron Ramsdale
    'RB': 'White',          # Ben White
    'CB1': 'Saliba',        # William Saliba
    'CB2': 'Gabriel',       # Gabriel Magalhães
    'LB': 'Zinchenko',      # Oleksandr Zinchenko
    'CM1': 'Ødegaard',      # Martin Ødegaard
    'CM2': 'Rice',          # Declan Rice
    'CM3': 'Havertz',       # Kai Havertz
    'RW': 'Saka',           # Bukayo Saka
    'ST': 'Jesus',          # Gabriel Jesus
    'LW': 'Martinelli'      # Gabriel Martinelli
}

for position, (x, y) in your_team.items():
    ax.text(x, y + 5, depth_labels_your_team[position], color='black', fontsize=10, ha='center', va='bottom')

# Add depth chart labels for the opposite team (generic placeholder)
depth_labels_opposite_team = {
    'GK': 'Player L',
    'RB': 'Player M',
    'CB1': 'Player N',
    'CB2': 'Player O',
    'LB': 'Player P',
    'CM1': 'Player Q',
    'CM2': 'Player R',
    'CM3': 'Player S',
    'RW': 'Player T',
    'ST': 'Player U',
    'LW': 'Player V'
}

for position, (x, y) in opposite_team.items():
    ax.text(x, y + 5, depth_labels_opposite_team[position], color='black', fontsize=10, ha='center', va='bottom')

# Add a legend to distinguish between Arsenal and the opposite team
handles, labels = ax.get_legend_handles_labels()
by_label = dict(zip(labels, handles))  # Remove duplicate labels
ax.legend(by_label.values(), by_label.keys(), loc='upper right', fontsize=10)

# Display the chart
plt.title('4-3-3 Formation: Arsenal vs Opposite Team', fontsize=14, pad=20)
plt.show()
```
