```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import time

def team_fixture_data(url, team_name, season):
    """
    Modified version of your original function with team_name and season parameters
    """
    try:
        page = urlopen(url).read()
        soup = BeautifulSoup(page, 'html.parser')
        count = 0 
        table = soup.find("tbody")
        
        if not table:
            print(f"No table found for {team_name} {season}")
            return pd.DataFrame()

        pre_df = dict()
        features_wanted = {"date", "time", "comp", "Round", "dayofweek", "venue", "result", 
                          "goals_for", "goals_against", "opponent", "xg_for", "xg_against", 
                          "possession", "attendance", "captain", "formation", "referee"}
        
        rows = table.find_all('tr')
        for row in rows:
            for f in features_wanted:
                if (row.find('th', {"scope":"row"}) != None) & (row.find("td",{"data-stat": f}) != None):
                    cell = row.find("td",{"data-stat": f})
                    a = cell.text.strip().encode()
                    text = a.decode("utf-8")
                    if f in pre_df:
                        pre_df[f].append(text)
                    else:
                        pre_df[f] = [text]
        
        if pre_df:
            df = pd.DataFrame.from_dict(pre_df)
            # Add team name and season columns
            df['team'] = team_name
            df['season'] = season
            return df
        else:
            print(f"No data found for {team_name} {season}")
            return pd.DataFrame()
            
    except Exception as e:
        print(f"Error fetching data for {team_name} {season}: {str(e)}")
        return pd.DataFrame()

def get_all_premier_league_data():
    """
    Fetch historic data for all Premier League top 6 teams from 2018-2019 to 2024-2025
    """
    # Team information: (team_code, team_name)
    teams = {
        '18bb7c10': 'Arsenal',
        '19538871': 'Manchester-United', 
        'cff3d9bb': 'Chelsea',
        '822bd0ba': 'Liverpool',
        'b8fd03ef': 'Manchester-City',  # You'll need to verify this code
        '361ca564': 'Tottenham-Hotspur'
    }
    
    # Seasons from 2018-2019 to 2024-2025
    seasons = [
        '2018-2019', '2019-2020', '2020-2021', '2021-2022', 
        '2022-2023', '2023-2024', '2024-2025'
    ]
    
    all_data = []
    
    for team_code, team_name in teams.items():
        print(f"Fetching data for {team_name}...")
        
        for season in seasons:
            print(f"  Season: {season}")
            
            # Construct URL in Arsenal format
            url = f"https://fbref.com/en/squads/{team_code}/{season}/matchlogs/c9/schedule/{team_name}-Scores-and-Fixtures-Premier-League"
            
            # Fetch data for this team and season
            df = team_fixture_data(url, team_name, season)
            
            if not df.empty:
                # Filter for Premier League matches with captain data (as in your original code)
                df_filtered = df.loc[(df['captain'] != '') ]
                if not df_filtered.empty:
                    all_data.append(df_filtered)
                    print(f"    Added {len(df_filtered)} matches")
                else:
                    print(f"    No Premier League matches with captain data found")
            else:
                print(f"    No data retrieved")
            
            # Add a small delay to be respectful to the website
            time.sleep(15)
    
    # Combine all data
    if all_data:
        combined_df = pd.concat(all_data, ignore_index=True)
        print(f"\nTotal matches collected: {len(combined_df)}")
        return combined_df
    else:
        print("No data collected")
        return pd.DataFrame()

# Usage example:
if __name__ == "__main__":
    # Get all historic data
    historic_data = get_all_premier_league_data()
    
    # Display summary
    if not historic_data.empty:
        print("\nData Summary:")
        print(f"Total matches: {len(historic_data)}")
        print(f"Teams: {historic_data['team'].unique()}")
        print(f"Seasons: {historic_data['season'].unique()}")
        print(f"Columns: {list(historic_data.columns)}")
        
        # Show sample data
        print("\nSample data:")
        print(historic_data.head())
        
        # You can save to CSV if needed
        # historic_data.to_csv('premier_league_top6_historic_data.csv', index=False)
    else:
        print("No data was collected successfully")

```

```python

# Alternative: If you want to fetch data for specific teams/seasons only
def get_specific_team_data(team_codes_dict, seasons_list):
    """
    Fetch data for specific teams and seasons
    
    Args:
        team_codes_dict: Dictionary with team_code: team_name pairs
        seasons_list: List of seasons in 'YYYY-YYYY' format
    """
    all_data = []
    
    for team_code, team_name in team_codes_dict.items():
        for season in seasons_list:
            url = f"https://fbref.com/en/squads/{team_code}/{season}/matchlogs/c9/schedule/{team_name}-Scores-and-Fixtures-Premier-League"
            df = team_fixture_data(url, team_name, season)
            
            if not df.empty:
                # df_filtered = df.loc[(df['captain'] != '') & (df['comp'] == 'Premier League')]
                df_filtered = df
                if not df_filtered.empty:
                    all_data.append(df_filtered)
    
    if all_data:
        return pd.concat(all_data, ignore_index=True)
    else:
        return pd.DataFrame()
```
