```python
import time
def league_performance_df(match_links):
    data_append = []
    for x in match_links:
        print(x)
        warnings.filterwarnings("ignore")
        url = x
        page =requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        name = [element.text for element in soup.find_all("span")]
        name = name[7]
        html_content = requests.get(url).text.replace('<!--', '').replace('-->', '')
        df = pd.read_html(html_content)
        df[0].columns = df[0].columns.droplevel(0) # drop top header row
        stats = df[0]
        stats = stats[(stats.Comp.isin(['La Liga', 'Premier League', 'Bundesliga', 'Serie A', 'Ligue 1'])) & (stats.Pos != "On matchday squad, but did not play")]
        season = stats[['Date','Gls', 'Ast',  'xG', 'npxG', 'xAG', 'SoT' ,'Squad']]
        columns_to_convert = ['Gls', 'Ast', 'xG', 'npxG', 'SoT' ,'xAG']
        for col in columns_to_convert:
            season[col] = pd.to_numeric(season[col], errors='coerce').fillna(0).astype(float)
        season = season.rename({'Squad': 'team'}, axis=1)
        season['Player'] = name
        data_append.append(season)
        del df, soup
        time.sleep(10)
    df_total = pd.concat(data_append)

    return df_total
```

```python
timeseries = league_performance_df(match_data_urls)
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# 1. Prepare the dataset
features = ["xG", "SoT",'xAG']

# Filter relevant columns
X = timeseries[features]
y_goals = timeseries["Gls"]
y_assists = timeseries["Ast"]

# Split into training & testing sets
X_train, X_test, y_train_goals, y_test_goals = train_test_split(X, y_goals, test_size=0.2, random_state=42)
X_train, X_test, y_train_assists, y_test_assists = train_test_split(X, y_assists, test_size=0.2, random_state=42)

# 2. Train Random Forest models
goals_model = RandomForestRegressor(n_estimators=100, random_state=42)
goals_model.fit(X_train, y_train_goals)

assists_model = RandomForestRegressor(n_estimators=100, random_state=42)
assists_model.fit(X_train, y_train_assists)

# 3. Evaluate the models (optional)
goal_preds = goals_model.predict(X_test)
assist_preds = assists_model.predict(X_test)

print("Goal Prediction MAE:", mean_absolute_error(y_test_goals, goal_preds))
print("Assist Prediction MAE:", mean_absolute_error(y_test_assists, assist_preds))

# 4. Predict Future Goals and Assists
timeseries["Predicted_Goals"] = goals_model.predict(X)
timeseries["Predicted_Assists"] = assists_model.predict(X)

# 5. Calculate Per-Game Stats
# Count the number of unique match dates for each player
timeseries['unique_games'] = timeseries.groupby('Player')['Date'].transform('nunique')

# Calculate predicted goals and assists per game
timeseries['Predicted_Goals_Per_Game'] = timeseries['Predicted_Goals'] / timeseries['unique_games']
timeseries['Predicted_Assists_Per_Game'] = timeseries['Predicted_Assists'] / timeseries['unique_games']

# 6. Extrapolate for 38 games assuming 90% availability
availability_factor = 0.9  # 90% availability
games_in_season = 38

# Extrapolate goals to a full season
timeseries['Extrapolated_Goals'] = timeseries['Predicted_Goals_Per_Game'] * games_in_season * availability_factor
timeseries['Extrapolated_Assists'] = timeseries['Predicted_Assists_Per_Game'] * games_in_season * availability_factor

# 7. Aggregate Player Contributions (per-game and extrapolated for season)
player_predictions = timeseries.groupby("Player")[["Predicted_Goals_Per_Game", "Predicted_Assists_Per_Game", "Extrapolated_Goals", "Extrapolated_Assists"]].sum()

# Display the predictions
player_predictions.reset_index(inplace=True)
```

```python
def sim_goal_on_xG(xG):
    # Ensure that xG is between 0 and 1
    xG = max(0, min(1, xG))  # Clip xG to be within [0, 1]
    goal = np.random.choice([0, 1], p=[1 - xG, xG])
    return goal

# Simulate goals for a player
def sim_player_goals(player_name, simulations=1000, data=None):
    df = data.copy()
    player_df = df[df['Player'] == player_name]  # Filter by player
    simulation_track = {'goals': [], 'simulation_id': []}
    
    for sim in range(0, simulations):
        total_goals = 0
        # Simulate goals based on xG values for each match
        for x in player_df['xG']:
            goal = sim_goal_on_xG(x)
            total_goals += goal
        simulation_track['goals'].append(total_goals)
        simulation_track['simulation_id'].append(sim)
    
    simulation_track_df = pd.DataFrame(simulation_track)
    return simulation_track_df

# Example usage: simulate goals for Alexander Isak
isak_simulation = sim_player_goals(player_name="Alexander Isak", simulations=5000, data=timeseries)
isak_simulation = isak_simulation.groupby(['goals']).count().reset_index()
isak_simulation['prob'] = isak_simulation['simulation_id'] / 5000

print(isak_simulation)
```

```python
import pandas as pd

# Initialize an empty DataFrame to store all the simulated results
df_simulated = pd.DataFrame()

# Get unique players from the timeseries
unique_players = timeseries['Player'].unique()

# Run simulation for each player
for player in unique_players:
    # Get the number of unique games for the player
    unique_games = timeseries[timeseries['Player'] == player]['unique_games'].iloc[0]
    
    # Run the simulation
    df_aux = sim_player_goals(player_name=player, simulations=5000, data=timeseries)
    
    # Scale the goals to a 38-game season
    df_aux['goals'] = (df_aux['goals'] * (35 / unique_games))
    # .round().astype(int)
    
    # Group by goals and calculate probabilities
    df_aux = df_aux.groupby(['goals']).count().reset_index()
    df_aux['prob'] = df_aux['simulation_id'] / 5000
    df_aux['Player'] = player  # Add the player column to track results for each player
    
    # Append to the main DataFrame
    df_simulated = pd.concat([df_simulated, df_aux])

# Reset the index of the final DataFrame
df_simulated = df_simulated.reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Set up plot aesthetics
plt.style.use('fivethirtyeight')  # Use FiveThirtyEight style


def plot_goal_sim(ax, player_name, label_x=False, label_y=False):
    ax.set_xlim(df_simulated['goals'].min(), df_simulated['goals'].max() - 3)
    ax.set_ylim(0, .3)

    # Filter the DataFrame for the specific player
    player_sim = df_simulated[df_simulated['Player'] == player_name]

    # Plot the bars in blue
    bars_ = ax.bar(player_sim['goals'], player_sim['prob'], color='#1f77b4', alpha=0.8)
    
    # Highlight the bar with the highest probability
    max_prob_index = player_sim['prob'].idxmax()  # Index of the bar with the highest probability
    max_prob_pos = player_sim.index.get_loc(max_prob_index)  # Get positional index for bars_
    bars_[max_prob_pos].set_color('#006400')  # Highlight in green
    predicted_goals = int(player_sim.loc[max_prob_index, 'goals'])  # Predicted goals (rounded to int)

    # Set up axis ticks and labels
    ax.yaxis.set_major_locator(ticker.MultipleLocator(.05))
    ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
    ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:.0f}'))
    ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:.0%}'))
    
    # Reduce the size of the axis tick labels
    ax.tick_params(axis='both', which='major', labelsize=8)
    
    ax.grid(False)

    
    if label_x:
        ax.set_xlabel('Goals', color='black', size=10)
    if label_y:
        ax.set_ylabel('Probability', color='black', size=10)

    # Add player title with predicted goals
    ax.text(
        x=0, y=ax.get_ylim()[1] * 1.1,
        s=f"{player_name}| Predicted Goals: {predicted_goals}",
        size=10,
        color='black',
        ha='left',
        va='center',
        weight='bold'
    )

# Create the plot
fig = plt.figure(figsize=(8, 4), dpi=200)
ax = plt.subplot()

# Call the function with the player's name
plot_goal_sim(ax, player_name='Benjamin Šeško', label_x=True, label_y=True)

plt.tight_layout()
plt.show()
```
